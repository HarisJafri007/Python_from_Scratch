```python
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
file_path = '/kaggle/input/data-visualizatiion/insurance.csv'
insurance_data = pd.read_csv(file_path)
insurance_data.shape
```

```python
insurance_data.head()
```

```python
insurance_data.tail()
```

```python
plt.figure(figsize=(8,6))
plt.title('Relation between BMI and Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
sns.scatterplot(data=insurance_data, x=insurance_data['bmi'], y=insurance_data['charges'])
plt.show()
```

```python
#get an idea of how many of these are smokers and nonsmokers
plt.figure(figsize=(8,6))
plt.title('Relation between BMI and Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
sns.scatterplot(data=insurance_data, x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
plt.show()
```

```python
#find out the correlation overall
plt.figure(figsize=(8,6))
plt.title('Relation between BMI and Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
sns.regplot(data=insurance_data, x=insurance_data['bmi'], y=insurance_data['charges'])
plt.show()
```

```python
#find out the correlation for smokers and non smokers seperately
plt.figure(figsize=(8,6))
sns.lmplot(data=insurance_data, x='bmi', y='charges', hue='smoker')
plt.title('Relation between BMI and Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
```

```python
#what about comparing a categorical data with numerical data
#find out the correlation for smokers and non smokers seperately
plt.figure(figsize=(8,6))
sns.swarmplot(data=insurance_data, x='smoker', y='charges', size=2)
plt.title('Relation between Smoker and Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges')
```
